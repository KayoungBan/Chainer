   ## Super class ##

```python
class hoge_maker_class:
    def __init__(self,n):
        print("object is materialized!")
        self.n = n
        print("n = {} hoge will be generated".format(self.n))
        return
    def print_some(self):
        for i in range(self.n):
            print("hoge {0}/{1}".format(i+1,self.n))
            
obj = hoge_maker_class(2)
obj.print_some()
```

```python
class fuga_maker_class(hoge_maker_class):
    def print_some(self):
        for i in range(self.n):
            print("fuge {0}/{1}".format(i+1,self.n))
            
obj = fuga_maker_class(2)
obj.print_some()
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,30,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,2,1)

ax.plot(x, y1, label = "sine", color = "red")
ax.plot(x, y2, label = "cosine", color = "black")

ax.set_xlim(0,30)
ax.set_ylim(-2,2)
ax.legend(loc = "best")
plt.grid()

ay = fig.add_subplot(2,2,2)
ay.plot(x, y1+y2)
plt.grid()

az = fig.add_subplot(2,2,3)
az.plot(x, -y1)
plt.grid()

aw = fig.add_subplot(2,2,4)
aw.plot(x, -y2)


plt.grid()
plt.show()
```
